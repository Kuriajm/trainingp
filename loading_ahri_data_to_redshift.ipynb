2. Generate txt files files for AHRI_CDM schema

```python
#The script for generating the text files


#Import pandas and psycopg
import pandas as pd
import psycopg2

#Define folder where i want the text to go
folder_path = r'D:\APHRC\Projects\ahri\ahri_cdm'

# PostgreSQL database connection parameters
db_params = {
    'host': 'localhost',
    'database': 'test',
    'user': 'postgres',
    'password': 'motomoto',
    'port': '5433'
}

# Connect to the PostgreSQL database
try:
    connection = psycopg2.connect(**db_params)
    cursor = connection.cursor()
    print("Connected to the database")
except (Exception, psycopg2.Error) as error:
    print("Error while connecting to the database:", error)
    exit()

# List of tables to export
#tables_to_export = ['person', 'condition_occurrence']
schema = 'ahri_cdm'

#,'concept_relationship' 'concept_ancestor'
query = f"SELECT table_name FROM information_schema.tables WHERE table_schema = '{schema}'"

try:
    cursor.execute(query)
    table_records = cursor.fetchall()
    table_names = [record[0] for record in table_records]

    print(f"Tables in schema '{schema}':")
    for table_name in table_names:
        print(table_name)
except (Exception, psycopg2.Error) as error:
    print("Error retrieving tables:", error)

tables_to_export = table_names


# Chunk size for data retrieval
chunk_size = 500000  # Adjust this value as needed

# Loop through tables and export in chunks
for table_name in tables_to_export:
    query = f"SELECT * FROM {schema}.{table_name}"

    try:
        # Create a generator to fetch data in chunks
        chunks = pd.read_sql_query(query, connection, chunksize=chunk_size)

        # Save each chunk as a TXT file
        chunk_num = 1
        dfs = []
        for chunk in chunks:
            print("We are at index : {}".format(chunk_num))
            dfs.append(chunk)
            concatenated_df = pd.concat(dfs, ignore_index=True)
            chunk_num += 1
        if not concatenated_df.empty:
            concatenated_df.to_csv("{}/{}.txt".format(folder_path,table_name), sep='\t', index=False)
    except (Exception, psycopg2.Error) as error:
        print(f"Error exporting {table_name}:", error)

# Close the cursor and connection
cursor.close()
connection.close()
print("Database connection closed")
```

5. Copying the data from s3 to redshift tables


```python
import boto3
import os

# Define your AWS credentials and Redshift details
aws_access_key_id = 'AKIA6FZBJPZNQALQKDGN'
aws_secret_access_key = 'HUywToz8vpt9fJ1S/IASVkiXGsytAM9wPeMrI++6'
redshift_iam_role_arn = 'arn:aws:iam::974490533467:role/mubas'
redshift_cluster_endpoint = 'your-redshift-cluster-endpoint'
redshift_database = 'your_database'
redshift_user = 'your_redshift_user'
redshift_password = 'your_redshift_password'

# Create an S3 client
s3 = boto3.client('s3', aws_access_key_id=aws_access_key_id, aws_secret_access_key=aws_secret_access_key)

# List all files in the S3 bucket
bucket_name = 'upload-data-redshift'
files = s3.list_objects(Bucket=bucket_name)['Contents']

# Create a Redshift connection
import psycopg2

conn = psycopg2.connect(
    dbname=redshift_database,
    user=redshift_user,
    password=redshift_password,
    host=redshift_cluster_endpoint,
    port=5439  # Update with your Redshift port
)

# Iterate over files and copy data to Redshift tables
for file_info in files:
    file_key = file_info['Key']
    table_name = file_key.split('/')[-1].split('.')[0]  # Assuming the file name is used as the table name
    s3_path = f's3://{bucket_name}/{file_key}'

    # Construct and execute COPY command
    copy_command = f"""
    COPY schema_name.{table_name}
    FROM '{s3_path}'
    IAM_ROLE '{redshift_iam_role_arn}'
    DELIMITER '\t' CSV
    IGNOREHEADER 1
    EMPTYASNULL
    BLANKSASNULL
    REGION 'us-east-1';
    """

    with conn.cursor() as cursor:
        cursor.execute(copy_command)

# Close the Redshift connection
conn.close()
```
